```python
def draw_scatter_hist(x,y,date, alpha):
  fig = plt.figure(figsize=(14,6))

  ax1 = fig.add_subplot(121)
  ax2 = fig.add_subplot(122)

  ax1.scatter(date[x], date[y], alpha=alpha)
  ax1.set_xlabel(x)
  ax1.set_ylabel(f"{y} concentration (ppm)")
  ax1.set_title(f"Correlation between {x} and {y}")

  ax2.hist(date[x]);
  ax2.set_xlabel(f"{x} Values")
  ax2.set_ylabel(f"Number of {x} Values")
  ax2.set_title(f"Histogram of {x}");

def draw_scatter(x,y,date, alpha):
  # plt.figure(figsize=(18,10))
  plt.scatter(date[x], date[y], alpha=alpha)
  plt.xlabel(x)
  plt.ylabel(f"{y} concentration (ppm)")
  plt.title(f"Correlation between {x} and {y}")
  # plt.xticks(rotation=90)
  plt.show()
```

### ScreenResolution

```python
df['ScreenResolution']
```

extract resolution from ScreenResoltuin and create new feature(resolution)

```python
df['resolution'] =df['ScreenResolution'].str.extract(r'(\d+x\d+)')
df['resolution']
```

extract screen type from ScreenResolution

```python
df['screentype'] = df['ScreenResolution'].str.replace(r'(\d+x\d+)','',regex=True)
df['screentype'] = df['screentype'].replace(r'(Full HD|Quad HD|quad hd|\+|/|4K Ultra HD)','',regex=True)
df['screentype']
```

extraction of having touchscreen from screen type

```python
df['touchscreen'] = df['screentype'].str.extract(r'(Touchscreen)')
df['touchscreen']
```

remove touchscreen in screentype feature

```python
df['screentype'] = df['screentype'].str.replace('Touchscreen', '')
df['screentype']
```

convert touchscreenfeature to numreical(0 or 1)

```python
df['touchscreen'] = df['touchscreen'].replace('Touchscreen', 1)
df['touchscreen'] = df['touchscreen'].replace(np.nan, 0)
df['touchscreen']
```

```python
df['screentype'].value_counts()
```

remove space and replace special char with nan in screentype feature

```python
df['screentype'] = df['screentype'].replace(r' ', '')
df['screentype'] = df['screentype'].replace(r'^\s*$', np.nan, regex=True)
df['screentype']
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
new_df['Company'] = label_encoder.fit_transform(df['Company'])
new_df['resolution'] = label_encoder.fit_transform(df['resolution'])
new_df['storage_1_type'] = label_encoder.fit_transform(df['storage_1_type'])
new_df['cpu_'] = label_encoder.fit_transform(df['cpu_'])
new_df['Gpu'] = label_encoder.fit_transform(df['Gpu'])
new_df['Product'] = label_encoder.fit_transform(df['Product'])
new_df['OpSys'] = label_encoder.fit_transform(df['OpSys'])
# new_df['screentype'] = label_encoder.fit_transform(df['screentype'])
#new_df['storage_2_type'] = label_encoder.fit_transform(df['storage_2_type'])
```

```python
x = new_df.drop(['Price_euros'],axis=1)
y = new_df['Price_euros']
```

```python
# kf = KFold(n_splits=10, shuffle=True, random_state=0)
# degrees = np.arange(1, 4)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
mse_cv = []
for degree in range(1,6):
  lr_model = LinearRegression()
  poly = PolynomialFeatures(degree)
  x_train_scaled_poly = poly.fit_transform(x_train_scaled)
  scores = cross_val_score(lr_model, x_train_scaled_poly, y_train, cv=10, scoring='neg_mean_squared_error')
  mse_cv.append(scores.mean())
```

```python
mse_cv = np.array(mse_cv) * -1
mse_cv
```

```python
plt.figure(figsize=(10,5))
plt.yscale('log')
plt.plot(range(1,6), mse_cv)
plt.xlabel("Degree")
plt.ylabel("MSE")
```

```python
poly = PolynomialFeatures(1)
x_train_scaled_poly = poly.fit_transform(x_train_scaled)
x_test_scaled_poly = poly.transform(x_test_scaled)
model = LinearRegression()
model.fit(x_train_scaled_poly, y_train)
y_pred = model.predict(x_test_scaled_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
